```python
#prep plots
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
sns.set(font_scale=1.5, font='DejaVu Sans')
import numpy as np
```

```python
#execution time vs particles_per_streamline
particles_per_streamline = np.array([41.0, 61.0, 121.0, 241.0])
execution_time_minutes = np.array([566.54, 523.37, 587.85, 635.74])
```

```python
#plot particles_per_streamline vs time
x = particles_per_streamline
y = execution_time_minutes/60.0
f, ax = plt.subplots(figsize=(10, 6))
p = ax.plot(x, y, marker='o', linestyle='-', linewidth=1, label='execution_time')
y_fit =  y[2]*(x/x[2])
p = ax.plot(x, y_fit, linestyle='dotted', linewidth=2, label=r'$\propto$ particles_per_streamline')
p = ax.set_title('particles_per_streamline versus execution_time')
p = ax.set_ylabel('execution_time    (hours)')
p = ax.set_xlabel('particles_per_streamline')
p = ax.legend()
```
